<h2> Dice </h2>

```python
dice1 = np.random.randint(1,7, size=(1,7,1000))
dice2 = np.random.randint(1,7, size=(1,7,1000))
```

```python
dice1
```

```python
sum = dice1 + dice2
sum
```

```python
probability_of_seven = np.bincount(sum.flatten())[7] / 1000
print(probability_of_seven)
```

```python
probability_of_seven = np.bincount(sum.flatten())[7] / 1000
print(probability_of_seven * 100)
```

Probility of 7 sum appering

```python
count_7 = np.count_nonzero(sum == 7)
print("Total times 7 appeared:", count_7)
```

```python
prob7 = count_7/len(sum)
```

```python
prob7
```

```python
doubles = dice1 == dice2
prob_doubles = np.count_nonzero(doubles) / len(dice1)
print("Probability of getting a double:", prob_doubles)

(unique, counts) = np.unique(sum, return_counts=True)
most_freq_sum = unique[np.argmax(counts)]
print("Most frequent sum:", most_freq_sum)
```

```python
import matplotlib.pyplot as plt

plt.hist(sum, bins=range(2, 14), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title('Frequency of Dice Sums (1000 Rolls)')
plt.xlabel('Sum')
plt.ylabel('Frequency')
plt.xticks(range(2, 13))
plt.grid(True)
plt.show()
```
